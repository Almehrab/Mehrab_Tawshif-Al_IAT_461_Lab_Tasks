### **Part 7: Mapping and Transformation** ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
df["Sex_Code"] = df["Sex"].map({"male": 0, "female": 1})
df[["Sex", "Sex_Code"]]
```

### Explanation
 `map()` function replaces categorical gender values with numeric codes.  
example, `male` is converted to `0` and `female` is converted to `1`

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

# Create age category column
df["Age_Group"] = df["Age"].apply(age_group)

df[["Age", "Age_Group"]]
```

### Explanation
a custom function classify passengers into age groups.  
 `apply()` function applies the `age_group()` function to each value in the `Age` column.

```python
import csv
import time
pure_data = df.to_dict(orient="records")
```

```python
start = time.time()
python_names = []
for row in pure_data:
    python_names.append(row["Name"])
end = time.time()
print("Pure Python time:", end - start)
python_names[:10]
```

### Explanation
loops through each dictionary in the dataset and appends the `Name` value into a new list

```python
start = time.time()
python_age_above_60 = []
for row in pure_data:
    if row["Age"] == row["Age"]:  
        if row["Age"] > 60:
            python_age_above_60.append(row)
end = time.time()
print("Pure Python time:", end - start)
python_age_above_60[:10]
```

### Explanation
 filters passengers whose age is greater than 60 using loops and conditional statements

```python
start = time.time()
python_fares_euro = []
for row in pure_data:
    fare_euro = row["Fare"] * 0.92
    python_fares_euro.append(fare_euro)
end = time.time()
print("Pure Python time:", end - start)
python_fares_euro[:10]
```
